```python
a = int(input("輸入身高(單位：公分)："))
b = int(input("輸入體重(單位：公斤)："))

a/=100
O = b/(a*a)
print("BMI:",O)
if O<18.5:
  print('體位:過輕')
elif 18.5<O<24 or O==18.5:
  print('體位:適中')
elif 24<O<27 or O==24:
  print('體位:過重')
else:
  print('體位:肥胖')
```

```python
a = int(input("請輸入第一個數字："))
b = input("請輸入運算符號：")
c = int(input("請輸入第二個數字："))
if b=='+':
  print(a+c)
elif b=='-':
  print(a-c)
elif b=='*':
  print(a*c)
elif b=='/':
  print(a/c)
```
